# Introduction
Random processes are very common in physics. When dealing with large numbers of molecules, such as in gases, we donâ€™t describe the trajectory of every single particle ($10^{23}$ particles is too much to handle) but often use probabilities to determine the average behaviour. Here, we look at some illustrative examples of using random numbers in physical applications.

## Random walks
The random walk problem is a classic example which, despite its simplicity, describes very well the basic physical phenomenon of _diffusion_. Diffusion is a process where par- ticles which are initially concentrated in a region in space spread out with time. For example, a drop of ink in water will gradually spread to the whole volume of water, even if the water is perfectly still. This is because the individual molecules (both the ink and the water) (i) have random directions and speeds, which depend on the temperature, and (ii) also collide with each other. It turns out that these two ingredients can be modelled very easily.

## The method
We first look at a random walk in 1-D for a single particle. We start the particle off at the origin of the x-axis and assume that it moves unit distance in each time step (the molecular speed is thus 1 unit). At the end of each time step we set the direction to left or right with equal probability, mimicking the effect of collisions. You can easily write the python code, but for flexibility of later python coding, I show here how to define a function, called `walk(steps)`, which is passed a single number steps and returns a vector of that length, which contains all the positions of the random walk.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x
````

IMPORTANT: Make sure you understand how this function is used, and how it works. The rest of this worksheet is based on using it. You can also use the vectorised version below.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    Vectorised method.
    '''
    x = np.random.choice([-1, 1], size=steps) # random -1 or 1
    x = np.cumsum(x) # cumulative sum
    return x
````

We also define _for each step_ a "mean squared position": the average of the squares of the positions of all walkers (**not** the square of the average) at each step. Those two quantities, the "mean position" and the "mean squared position" are standard quantities in random walks.
````python
number_steps = 1000  # number of steps each walker will take
number_particles = 10  # the number of walkers to simulate
particles = np.zeros((number_particles, number_steps), int) #defines an empty array that will be filled later with array values

def walk(steps):
    x = np.zeros(steps, int)
    x[0] = 0
    for i in range(1, len(x)):
        rnd = random()
        if (rnd < 0.5):
            x[i] = x[i-1] + 1
        else:
            x[i] = x[i-1] - 1
    return x

for i in range(number_particles):
  particles[i, :]=walk(number_steps) #fills the particle array with the data generated from the for loop
  plt.plot(particles[i, :]) # plots every row that is stored in the particles array, up to the value for number_particles

#for i in range(number_particles):

plt.xlabel('Step Number')
plt.ylabel('Position')
plt.title(f'Plot of {number_particles} Random Walks')
#removed the legend for this specific plot, as it made the graph hard to view
````

````python
last_step_position=particles[:, -1] #gets the 10 entries for the last entry of the particles array (essentially the values of the position at step 1000)

last_step_mean=np.mean(last_step_position) #mean
last_step_mean_sq=np.mean(last_step_position**2) #mean squared
#these values change every time particles is reran

print(f"The mean of the positions at the last step is: {last_step_mean}, and the the value for the mean squared is {last_step_mean_sq}.")

step_mean_position=np.zeros(number_steps) #creates an empty array with identical dimensions to the array number_steps
step_mean_position_squared=np.zeros(number_steps)

for i in range(number_steps): #for loop for values in number_steps (loops for every value in this array)
  step_position=particles[:, i] #obtains every row for the loop, obtaining every step value at step i
  step_mean=np.mean(step_position) #mean
  step_mean_squared=np.mean(step_position**2)
  step_mean_position[i]=step_mean #fills the array, storing the mean position for step i
  step_mean_position_squared[i]=step_mean_squared

plt.subplot(2,1,1)
plt.plot(range(number_steps), step_mean_position, label='Mean Position')
plt.ylabel('Mean Position')
plt.title('Graph of the Mean Position at Each Step')
plt.legend(loc='best')

plt.subplot(2,1,2)
plt.plot(range(number_steps), step_mean_position_squared, label='Mean Position Squared')
plt.xlabel('Step Number')
plt.ylabel('Mean Position Squared')
plt.legend(loc='best')
````
